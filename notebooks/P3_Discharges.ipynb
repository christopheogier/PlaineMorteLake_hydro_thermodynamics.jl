## Discharges 

```julia
# Integration boundaries.  If the sensor was submerged before logging started, then
# you can probably delete t1s and t2s (also in below loop).  If not, like in the fake-examples below,
# the boundaries need to be set to avoid integrating the negative concentrations from 0-15s.
#t1s = [15, 15]
#t2s = [Inf, Inf]

# Store the times and discharges
ts3 = DateTime[]
Qs3 = []  # this will contain the discharges
expe3=[]
for i = 1:length(traces)
    Q = calcQ(traces[i],0,Inf,delta_cond2conc_lab_2000[traces[i][:sensor]]) #, t1s[i], t2s[i]) # in m^3/s
    expe=traces[i][:id]
    #if Q<1000
         println("Discharge for experiment $expe is $(round(Q*1000,sigdigits=3)) ± $(round(std(Q)*1000,sigdigits=3)) l/s")
    #else
       # println("Discharge for experiment $expe is $(round(Q,sigdigits=2)) m^3/s")
    #end
    push!(Qs3, Q)
    push!(ts3, traces[i][:tstart])
    push!(expe3, expe)
end;
```

```julia
fig=figure()
pygui(false)
# plot the discharge vs time:
scatter(ts3, mean.(Qs3))
#scatter(string.(ts), Qs) # might be better?
xlabel("injection time")
ylabel("discharge (m³/s)")
fig[:autofmt_xdate](bottom=0.2,rotation=30,ha="right")
xlim(Date(2019,7,4),Date(2019,8,4))
axvline(x=DateTime(2019,7,10,12),linewidth=1,color="r")
title("Discharges at P3")
grid(true)
```

## Calibration sensitivity 

```julia
ts3_200 = DateTime[]
Qs3_200 = []  # this will contain the discharges
expe1=[]
for i = 1:length(traces)
    Q = calcQ(traces[i],0,Inf,delta_cond2conc_lab_200[traces[i][:sensor]]) #, t1s[i], t2s[i]) # in m^3/s
    expe=traces[i][:id]
    if Q<1000
        println("Discharge for experiment $expe is $(round(Q*1000,sigdigits=2)) l/s")
    else
        println("Discharge for experiment $expe is $(round(Q,sigdigits=2)) m^3/s")
    end
    push!(Qs3_200, Q)
    push!(ts3_200, traces[i][:tstart])
    push!(expe1, expe)
end;
```

```julia
scatter(mean.(Qs3),mean.(Qs3_200))
xlabel("Q [0-2000]μS/cm")
ylabel("Q [0-200]μS/cm")
plot([0,1],[0,1],linestyle="-",color="black")
plot([0,1],[0,0.7],linestyle="-",color="blue")
text(0.6,0.3,"30%",color="blue")
grid(true)
```

## To store Discharge at P3 in CSV file 

```julia
using DataFrames
using CSV
```

```julia
q = round.(mean.(Qs3), digits = 3)
sigma_Qs3 = round.(std.(Qs3), digits = 3)
df=DataFrame(Date_Time = ts3, Discharge = q, Std_dev = sigma_Qs3 )
CSV.write(path_products*"discharge_P3.csv",df ; dateformat = "yyyy-mm-ddTH:M:S");
```
