## Discharges

```julia

# Integration boundaries.  If the sensor was submerged before logging started, then
# you can probably delete t1s and t2s (also in below loop).  If not, like in the fake-examples below,
# the boundaries need to be set to avoid integrating the negative concentrations from 0-15s.
#t1s = [15, 15]
#t2s = [Inf, Inf]

# Store the times and discharges
ts2 = DateTime[]
Qs2 = []  # this will contain the discharges
expe2=[]
for i = 1:length(traces)
    Q = calcQ(traces[i],0,Inf,delta_cond2conc_lab_2000[traces[i][:sensor]]) #, t1s[i], t2s[i]) # in m^3/s
    expe=traces[i][:id]
    #if Q<1000
        println("Discharge for experiment $expe is $(round(Q*1000,sigdigits=2)) ± $(round(std(Q)*1000,sigdigits=2)) l/s")
    #else
        #println("Discharge for experiment $expe is $(round(Q,sigdigits=2)) m^3/s")
    #end
    push!(Qs2, Q)
    push!(ts2, traces[i][:tstart])
    push!(expe2, expe)
end;
```

```julia
fig=figure()
pygui(false)
# plot the discharge vs time:
scatter(ts2, mean.(Qs2))
fig[:autofmt_xdate](bottom=0.2,rotation=30,ha="right")
xlabel("injection time")
ylabel("discharge (m³/s)")
xlim(Date(2019,7,4),Date(2019,8,4))
axvline(x=DateTime(2019,7,10,12),linewidth=1,color="r")
title("Discharges at P2")
grid(true)
```

## Calibration sensitivity 

```julia
ts2_200 = DateTime[]
Qs2_200 = []  # this will contain the discharges
expe1=[]
for i = 1:length(traces)
    Q = calcQ(traces[i],0,Inf,delta_cond2conc_lab_200[traces[i][:sensor]]) #, t1s[i], t2s[i]) # in m^3/s
    expe=traces[i][:id]
    push!(Qs2_200, Q)
    push!(ts2_200, traces[i][:tstart])
    push!(expe1, expe)
end;
```

```julia
scatter(mean.(Qs2),mean.(Qs2_200))
xlabel("Q [0-2000]μS/cm")
ylabel("Q [0-200]μS/cm")
plot([0,1],[0,1],linestyle="-",color="black")
plot([0,1],[0,1.1],linestyle="-",color="blue")
text(0.5,0.8,"10%",color="blue")
grid(true)
```

## To store Discharge at P2 in a CSV file

```julia
using DataFrames
using CSV
```

```julia
q = round.(mean.(Qs2), digits = 3)
sigma_Qs2 = round.(std.(Qs2), digits = 3)
df=DataFrame(Date_Time = ts2, Discharge = q, Std_dev = sigma_Qs2)
CSV.write(path_products*"discharge_P2.csv",df ; dateformat = "yyyy-mm-ddTH:M:S");
```
