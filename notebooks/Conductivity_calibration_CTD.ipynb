### To convert ml added to concentration 

```julia
using NBInclude
@nbinclude("helper-functions.ipynb");
```

```julia
"""
Converts ml added to bucket to a concentration (g/l == kg/m^3).

Input:

- ml -- how many mililiters were added
- solution -- the concentration of the calibration solution (kg/m^3 == g/l)
- bucketsize -- the size of the bucket/bottle to which the solution was added (l)

Output:

- concentration (kg/m^3 == g/l)
"""
function ml_to_concentration(ml, solution, bucketsize)
    mass = ml/1e3 * solution # salt mass added to bucket (g)
    return mass/bucketsize # concentration in g/l (== kg/m^3)
end
# For example, convert cali1[:,1] to concentration (g/l):
#ml_to_concentration(cali1_44[:,1], solution, bucketsize)
```

## Function `fit_calibration` to get concentration from conductivity readouts

What we are really after is a function which tells us the concentration for a given sensor readout.  This is done with the `fit_calibration` function which is contained in the extra file [`helper-functions.ipynb`](helper-functions.ipynb). Running this returns us just such a function, by fitting a straight line through the data:

$f(x) = ax $ for the calibration on field and the one in lab using [0-200]μS/cm range

$f(x) = ax^2 + bx $ for the one in lab using [0-2000]μS/cm range (it is the best fit according to me) 

where $x$ is difference between the readout and the readout at 0 concentration, and $a,b$ is the parameter to fit.
It will also tell you how good the fit is by giving the error on `a`. The error should be "reasonably" small.

```julia
@nbinclude("helper-functions.ipynb");

#calibration "classic" on field: 1 and 2
delta_cond2conc= Dict()   
delta_cond2conc[205144] = fit_calibration(bucketsize, solution2, cali1_44, cali2_44);
delta_cond2conc[205145] = fit_calibration(bucketsize, solution2, cali1_45, cali2_45);
delta_cond2conc[207265] = fit_calibration(bucketsize, solution2, cali1_65, cali2_65);
delta_cond2conc[205209] = fit_calibration(bucketsize, solution2, cali1_09, cali2_09);

#calibration 3rd
delta_cond2conc_lab_2000= Dict()
delta_cond2conc_lab_2000[205144] = fit_calibration_LAB_2000(cali3_44);
delta_cond2conc_lab_2000[205145] = fit_calibration_LAB_2000(cali3_45);
delta_cond2conc_lab_2000[207265] = fit_calibration_LAB_2000(cali3_65);
delta_cond2conc_lab_2000[205209] = fit_calibration_LAB_2000(cali3_09);

#calibration 4th
delta_cond2conc_lab_200= Dict()
delta_cond2conc_lab_200[205144] = fit_calibration_LAB_200(cali4_44);
delta_cond2conc_lab_200[205145] = fit_calibration_LAB_200(cali4_45);
delta_cond2conc_lab_200[207265] = fit_calibration_LAB_200(cali4_65);
delta_cond2conc_lab_200[205209] = fit_calibration_LAB_200(cali4_09);
```

## CTD 205144

```julia
pygui(false)
#calibration 3rd
conc1 = ml_to_concentration(cali3_44[1:10,1], solution1, bucketsize2);
conc2 = ml_to_concentration(cali3_44[11:19,1], solution2, bucketsize2) .+ conc1[end];
conc3=vcat(conc1,conc2);
delta_readout3 = cali3_44[:,2].-cali3_44[1,2];     #to subsracte 0 reading
#calibration 4th
conc4 = ml_to_concentration(cali4_44[1:10,1], solution1, bucketsize2);
conc5 = ml_to_concentration(cali4_44[11:20,1], solution2, bucketsize2) .+ conc1[end];
conc6=vcat(conc4,conc5);
delta_readout6 = cali4_44[:,2].-cali4_44[1,2];     #to subsracte 0 reading

## Plot the calibration points:

# scatter plots (x,y) points
fig = figure()
scatter(conc3,delta_readout3, 
        label="Calibration Lab [0-2000μS/cm]")
scatter(conc6,delta_readout6, 
        label="Calibration Lab [0-200μS/cm]")
scatter(ml_to_concentration(cali1_44[:,1], solution2, bucketsize), cali1_44[:,2].-cali1_44[1,2], 
        label="Calibrations on field", color="red")
scatter(ml_to_concentration(cali2_44[:,1], solution2, bucketsize), cali2_44[:,2].-cali2_44[1,2], color="red")

readouts = range(0,stop=400,length=100) #that's conductivity values (well, since we want conc from cond in discharge calculation, cond as x-axis would makes more sens)

# Now plot the line of best fit:
plot(delta_cond2conc[205144](readouts), readouts, color="red")
plot(mean.(delta_cond2conc_lab_2000[205144](readouts)), readouts,color="blue")
plot(delta_cond2conc_lab_200[205144](readouts), readouts, color="orange")

xlim((-0.001,0.3));
title("CTD 205144");
xlabel("concentration (g/l)")
ylabel("Sensor readout change (μS/cm)")
legend()
grid(true)

savefig(path_results_cali_cond*"205144.png")
```
